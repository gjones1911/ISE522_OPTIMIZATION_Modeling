```python
import gurobipy as gp
from gurobipy import GRB
!pip install xelatex
```

```python
try:

    # Create a new model
    m = gp.Model("IA2")

    # Create variables (amount to order each week)
    x1 = m.addVar(vtype=GRB.CONTINUOUS, name="x1")
    x2 = m.addVar(vtype=GRB.CONTINUOUS, name="x2")
    x3 = m.addVar(vtype=GRB.CONTINUOUS, name="x3")
    x4 = m.addVar(vtype=GRB.CONTINUOUS, name="x4")
    x5 = m.addVar(vtype=GRB.CONTINUOUS, name="x5")
    x6 = m.addVar(vtype=GRB.CONTINUOUS, name="x6")
    x7 = m.addVar(vtype=GRB.CONTINUOUS, name="x7")
    x8 = m.addVar(vtype=GRB.CONTINUOUS, name="x8")
    x9 = m.addVar(vtype=GRB.CONTINUOUS, name="x9")
    x10 = m.addVar(vtype=GRB.CONTINUOUS, name="x10")
    
    # Create variables (to order or not(1/0))
    o1 = m.addVar(vtype=GRB.BINARY, name="o1")
    o2 = m.addVar(vtype=GRB.BINARY, name="o2")
    o3 = m.addVar(vtype=GRB.BINARY, name="o3")
    o4 = m.addVar(vtype=GRB.BINARY, name="o4")
    o5 = m.addVar(vtype=GRB.BINARY, name="o5")
    o6 = m.addVar(vtype=GRB.BINARY, name="o6")
    o7 = m.addVar(vtype=GRB.BINARY, name="o7")
    o8 = m.addVar(vtype=GRB.BINARY, name="o8")
    o9 = m.addVar(vtype=GRB.BINARY, name="o9")
    o10 = m.addVar(vtype=GRB.BINARY, name="o10")
     
    # Create variables (demand of week w)
    d1 = m.addVar(vtype=GRB.CONTINUOUS,lb=10, ub=10, name="d1")
    d2 = m.addVar(vtype=GRB.CONTINUOUS,lb=10, ub=10, name="d2")
    d3 = m.addVar(vtype=GRB.CONTINUOUS,lb=10, ub=10, name="d3")
    d4 = m.addVar(vtype=GRB.CONTINUOUS,lb=0, ub=0, name="d4")
    d5 = m.addVar(vtype=GRB.CONTINUOUS,lb=0, ub=0, name="d5")
    d6 = m.addVar(vtype=GRB.CONTINUOUS,lb=15, ub=15, name="d6")
    d7 = m.addVar(vtype=GRB.CONTINUOUS,lb=20, ub=20, name="d7")
    d8 = m.addVar(vtype=GRB.CONTINUOUS,lb=20, ub=20, name="d8")
    d9 = m.addVar(vtype=GRB.CONTINUOUS,lb=0, ub=0, name="d9")
    d10 = m.addVar(vtype=GRB.CONTINUOUS,lb=10, ub=10, name="d10")
    
    # Create variables (demand of week w)
    h1 = m.addVar(vtype=GRB.CONTINUOUS, name="h1")
    h2 = m.addVar(vtype=GRB.CONTINUOUS, name="h2")
    h3 = m.addVar(vtype=GRB.CONTINUOUS, name="h3")
    h4 = m.addVar(vtype=GRB.CONTINUOUS, name="h4")
    h5 = m.addVar(vtype=GRB.CONTINUOUS, name="h5")
    h6 = m.addVar(vtype=GRB.CONTINUOUS, name="h6")
    h7 = m.addVar(vtype=GRB.CONTINUOUS, name="h7")
    h8 = m.addVar(vtype=GRB.CONTINUOUS, name="h8")
    h9 = m.addVar(vtype=GRB.CONTINUOUS, name="h9")
    h10 = m.addVar(vtype=GRB.CONTINUOUS, name="h10")
    
    # Set objective
    m.setObjective(((h1*2)+(100*o1))+((h2*2)+(100*o3))+((h3*2)+(100*o3))+
                   ((h4*2)+(100*o4))+((h5*2)+(100*o5))+((h6*2)+(100*o6))+
                   ((h7*2)+(100*o7))+((h8*2)+(100*o8))+((h9*2)+(100*o9))+
                   ((h10*2)+(100*o10)), GRB.MINIMIZE)

    # Add constraints
    m.addConstr(x1>=0, "c1")
    m.addConstr(x2>=0, "c2")
    m.addConstr(x3>=0, "c3")
    m.addConstr(x4>=0, "c4")
    m.addConstr(x5>=0, "c5")
    m.addConstr(x6>=0, "c6")
    m.addConstr(x7>=0, "c7")
    m.addConstr(x8>=0, "c8")
    m.addConstr(x9>=0, "c9")
    m.addConstr(x10>=0, "c10")
    
    # Add constraints
    m.addConstr(h1>=1, "c11")
    m.addConstr(h2>=1, "c12")
    m.addConstr(h3>=1, "c13")
    m.addConstr(h4>=0, "c14")
    m.addConstr(h5>=0, "c15")
    m.addConstr(h6>=1.5, "c16")
    m.addConstr(h7>=2, "c17")
    m.addConstr(h8>=2, "c18")
    m.addConstr(h9>=0, "c19")
    m.addConstr(h10>=1, "c20")


    #Add constraint:
    m.addConstr(h1==(x1*o1)-10+20, "c21")
    m.addConstr(h2==(x2*o2)-10+h1, "c22")
    m.addConstr(h3==(x3*o3)-10+h2, "c23")
    m.addConstr(h4==(x4*o4)-0+h3, "c24")
    m.addConstr(h5==(x5*o5)-0+h4, "c25")
    m.addConstr(h6==(x6*o6)-15+h5, "c26")
    m.addConstr(h7==(x7*o7)-20+h6, "c27")
    m.addConstr(h8==(x8*o8)-20+h7, "c28")
    m.addConstr(h9==(x9*o9)-0+h8, "c29")
    m.addConstr(h10==(x10*o10)-10+h9, "c30")

    
    # Optimize model
    m.optimize()

    for v in m.getVars():
        print('%s %g' % (v.VarName, v.X))

    print('Obj: %g' % m.ObjVal)

except gp.GurobiError as e:
    print('Error code ' + str(e.errno) + ': ' + str(e))

except AttributeError:
    print('Encountered an attribute error')
```
